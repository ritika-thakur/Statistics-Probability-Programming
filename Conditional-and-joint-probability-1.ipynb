#  Demonstration of Practical definition of Conditional and Joint Probability



Since this data is a sample (of dimensions 1000x6 ) not poulation, we are going to find relative frequency instead of probability which is very close (approximately equal) to probability.

```python
import numpy as np
```

```python
marks_of_students = np.random.randint(low=0,high=100,size=(1000,6))
```

```python
marks_of_students
```

```python
import pandas as pd
```

```python
marks_data = pd.DataFrame(data=marks_of_students, columns=['Python','DSA','COA','OS','CD','ML'])
```

```python
marks_data
```

Now to hire a student for ML, the Marks of Python, DSA and COA should be greater than or equal to 50, given that the marks in ML is greater than 50.

\begin{equation}
P(Python>=50 \cap DSA>=50 \cap COA>=50 | ML>50) = \frac{P(Python>=50  \cap  DSA>=50  \cap  COA>=50  \cap  ML>=50)}{P(ML > 50)}
\end{equation}

```python
good_marks_in_ML = marks_data[marks_data['ML'] > 50]
```

```python
good_marks_in_ML
```

```python
boolean_mask = (good_marks_in_ML['Python'] >= 50) & (good_marks_in_ML['DSA'] >= 50) & (good_marks_in_ML['COA'] >= 50)
```

```python
good_marks_in_py_dsa_coa = good_marks_in_ML[boolean_mask]
```

```python
good_marks_in_py_dsa_coa
```

```python
P_py_dsa_coa_on_ML = good_marks_in_py_dsa_coa.shape[0] / good_marks_in_ML.shape[0]
```

```python
P_py_dsa_coa_on_ML
```
