```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
level_of_education = ['Senior Secondary']*30 + ['Bachelors']*30
```

```python
level_of_education
```

```python
teaching_methodology = ['Theory']*20 + ['Hands On']*20 + ['Combination']*20
```

```python
teaching_methodology
```

```python
ss_theory = np.random.normal(loc=34, scale=2, size=10)

ss_handson = np.random.normal(loc=45, scale=4, size=10)

ss_combination = np.random.normal(loc=51, scale=3.5, size=10)

bachelors_theory = np.random.normal(loc=30, scale=1.5, size=10)

bachelors_handson = np.random.normal(loc=55, scale=6, size=10)

bachelors_combination = np.random.normal(loc=48, scale=2.5, size=10)

attention_span = np.concatenate((ss_theory, ss_handson, ss_combination, bachelors_theory, bachelors_handson, bachelors_combination))
```

```python
attention_span
```

```python
df['LE'] = level_of_education

df['TM'] = teaching_methodology

df['AS'] = attention_span
```

```python
df
```

```python
import statsmodels.api as sm
```

```python
from statsmodels.formula.api import ols
```

```python
model = ols(formula='AS ~ C(LE) + C(TM) + (C(LE)*C(TM))', data=df).fit()
```

```python
sm.stats.annova_lm(model, typ=2)
```
